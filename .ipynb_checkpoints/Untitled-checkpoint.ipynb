```python
import pandas as pd
import os
import glob
```

```python
df = pd.read_csv(r'f:\TUNEL\Measurement\Tunel_4.5Nsw.csv')
```

```python
df.head()
```

```python
os.chdir(r'f:\TUNEL')
df['Sample'] = None
df['Group'] = None
for idx, row in df.iterrows():
    f = row['Image']
    f = glob.glob('./Wasi/*/'+f)[0]
    f = os.path.dirname(f)
    sample = f[7:12]
    group = sample[3]
    if group == 'L':
        group = 'Loaded'
    elif group == 'R':
        group = 'Nonloaded'
        
    df['Sample'][idx] = sample
    df['Group'][idx] = group
```

```python
#df["Total"] = df['Neg'] + df['Pos']

dfag = df.groupby(['Sample', 'Group']).sum()
```

```python
dfag['Sum'] = dfag['Positive'] + dfag['Negative'] + dfag['Empty']
dfag['Tunel_rate'] = dfag['Positive'] / (dfag['Positive'] + dfag['Negative'])
dfag['Dying_rate'] = dfag['Positive'] / (dfag['Positive'] + dfag['Negative'] + dfag['Empty'])
dfag
```

```python
import pandas as pd
import os
import glob
```

```python
df = pd.read_csv(r'F:\Measurement-Endomucin\Endomucin_Results.csv')
```

```python
imgls = df['# ID'].values
```

```python
os.chdir(r'E:\DATA\4.5N\Endomucin')
```

```python
s= []
g =[]
for img in imgls:
    p = glob.glob(r'./*slice*/'+img+'.tif')
    p = p[0]
    p = os.path.dirname(p)
    fn = p[2:]
    sample = fn[:5]
    group = 'Loaded' if sample[4] == 'L' else 'Nonloaded'
    s.append(sample)
    g.append(group)
```

```python
df['Sample'] = s
df['Loading'] = g
```

```python
df.head()
```

```python
df2 = df.groupby('Sample').mean()
```

```python
df2.to_csv(r'C:\Users\wangs\Google Drive\4th batch bone mets 4.5 N statistical analysis and plotting\4.5N Endomucin stats.csv')
df.to_csv(r'C:\Users\wangs\Google Drive\4th batch bone mets 4.5 N statistical analysis and plotting\4.5N Endomucin.csv')
```
